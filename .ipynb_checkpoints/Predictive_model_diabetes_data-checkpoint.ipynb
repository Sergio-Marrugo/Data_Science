```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
# Our imput is the diabetes data table that we just cleaned and transformed with imputed null values.
input_file = "C:/Users/EQUIPO/Desktop/py4e/Diabetes_imputed.csv"
diabetes_data = pd.read_csv(input_file)
diabetes_data.head(5)
```

```python
#We run a preliminary logistic model with all the variables in r and selescted only the ones that are significant
X_features = ['Pregnancies','Glucose','BMI_na','DiabetesPedigreeFunction', 'Insulin'] #Insulin is marginally significant
X = diabetes_data[X_features]
```

```python
y = diabetes_data.Outcome
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
model = LogisticRegression()
```

```python
model.fit(train_X, train_y)
```

```python
y_pred = model.predict(val_X)
```

```python
model.score(val_X, val_y)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(val_y, y_pred))
```

```python
#classification report
from sklearn.metrics import classification_report
```

```python
print(classification_report(val_y, y_pred))
```
